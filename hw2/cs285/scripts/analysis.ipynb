```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pickle
import glob
# import tensorflow as tf
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

# Experiment 1

```python
def plot_comparison(exp, dir):

    # flags = [exp + i for i in ['no_rtg_dsa', 'rtg_dsa', 'rtg_na']]

    f = glob.glob(f"data/{dir}*")
    # print(dir, f)
    f.sort()
    plt.figure(figsize=(8,5))

    for i in f:
        data = EventAccumulator(i)
        data.Reload()
        # print(data.Tags())

        ''' load results '''

        time, step, Eval_AverageReturn = zip(*data.Scalars('Eval_AverageReturn'))

        Eval_AverageReturn = np.array(Eval_AverageReturn)

        ''' make plot '''

        plt.plot(step, Eval_AverageReturn, '-', label=f"{i.split('q2_pg_')[1].split('-v')[0]}")


        plt.legend()
        plt.xlabel('Iteration')
        plt.ylabel('Eval_AverageReturn')
        plt.title(f'{exp}')
        # plt.yscale('log')
    plt.show()
```

```python
plot_comparison('small batch', 'q2_pg_q1_sb')
```

# Experiment 2

```python
def plot_comparison(exp, dir):

    f = glob.glob(f"data/{dir}*")
    # print(dir, f)
    f.sort()
    plt.figure(figsize=(8,5))

    data = EventAccumulator(f[0])
    data.Reload()
    print(data.Tags())

    ''' load results '''

    time, step, Eval_AverageReturn = zip(*data.Scalars('Eval_AverageReturn'))
    _, _, Eval_StdReturn = zip(*data.Scalars('Eval_StdReturn'))
    Eval_AverageReturn = np.array(Eval_AverageReturn)

    ''' make plot '''

    plt.plot(step, Eval_AverageReturn, '-', label=f"{f[0].split('q2_pg_')[1].split('-v')[0]}")
    plt.fill_between(step, Eval_AverageReturn - Eval_StdReturn, Eval_AverageReturn + Eval_StdReturn, alpha=0.2)


    plt.legend()
    plt.xlabel('Iteration')
    plt.ylabel('Eval_AverageReturn')
    plt.title(f'{exp}')
    # plt.yscale('log')
    plt.show()
```

```python
plot_comparison('Inverted pendulum', 'q2_pg_q2_b200_r005_InvertedPendulum-v4_25-09-2022_15-29-56')
```

# Experiment 4

```python
def plot_comparison(exp, dir):

    # flags = [exp + i for i in ['no_rtg_dsa', 'rtg_dsa', 'rtg_na']]

    f = glob.glob(f"data/{dir}*")
    f.sort()
    plt.figure(figsize=(8,5))

    for i in f:
        data = EventAccumulator(i)
        data.Reload()
        ''' load results '''

        time, step, Eval_AverageReturn = zip(*data.Scalars('Eval_AverageReturn'))
        Eval_AverageReturn = np.array(Eval_AverageReturn)

        ''' make plot '''

        if 'b10000' in i:
            ls=':'
        if 'b30000' in i:
            ls='-.'
        if 'b50000' in i:
            ls='-'
        if 'lr0005' in i:
            c = 'dodgerblue'
        if 'lr001' in i:
            c = 'darkorange'
        if 'lr002' in i:
            c = 'orangered'
        plt.plot(step, Eval_AverageReturn, linestyle=ls, color=c, label=f"{i.split('search_')[1].split('_rtg')[0].split('-v')[0]}")

    plt.axhline(0,linestyle='--')
    plt.legend()
    plt.xlabel('Iteration')
    plt.ylabel('Eval_AverageReturn')
    plt.title(f'{exp}')
    # plt.yscale('log')
    plt.show()
```

```python
plot_comparison('Cheetah', 'q2_pg_q4_search')
```
